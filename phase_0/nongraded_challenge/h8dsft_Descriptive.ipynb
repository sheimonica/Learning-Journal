```python
import pandas as pd

data = pd.read_csv('covid_19_indonesia_time_series_all.csv')
data
```

### Cari Measure of Central Tendency dari data tersebut.

```python
# Get Numerical and Categorical Columns

import numpy as np

num_columns = data.select_dtypes(include=np.number).columns.tolist()
cat_columns = data.select_dtypes(include=['object']).columns.tolist()

print('Numeric Columns : ', num_columns)
print('Categorical Columns : ', cat_columns)
```

```python
# Get Mean, Median, and Mode

data_mean = data[num_columns].mean().rename('mean')
data_median = data[num_columns].median().rename('median')
data_mode = data[num_columns + cat_columns].mode().iloc[0].rename('mode')

central_tendency = pd.concat([data_mean, data_median, data_mode], axis=1)
central_tendency
```

### Cari Measure of Variance dari data tersebut.

```python
# Get Variance and Standard Deviation

data_variance = data[num_columns].var().rename('variance')
data_std = data[num_columns].std().rename('standard_deviation')

var_std = pd.concat([data_variance, data_std], axis=1)
var_std
```

```python
# Merge DataFrame of `central_tendency` and `var_std`

all_measurements = pd.concat([central_tendency, var_std], axis=1)
all_measurements
```
