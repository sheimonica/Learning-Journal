```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('covid_19_indonesia_time_series_all.csv')
data
```

```python
indo = data[
        (data["Location ISO Code"] == "IDN")] # Check if data is indonesia

indo
```

```python
case_rate = indo[['Date','Total Cases per Million']].groupby('Date').sum()
print('Average Total Cases per Million for a year: {}'.format(np.round(case_rate['Total Cases per Million'].mean())))
```

- H0 = Total Cases per Million <= 5000
- H1 = Total Cases per Million > 5000

```python
case_rate
```

```python
t_stat,p_val = stats.ttest_1samp(case_rate['Total Cases per Million'], 5000)
print('P-value:',p_val/2) #The p-value divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat)
```

**Since our p-value is less than `0.05`, so we reject the null hypothesis** and we can conclude that, Average Total Cases per Million for a years is bigger than 5000 Million.
